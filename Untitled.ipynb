```python
df.groupby('Category')['Review Rating'].median()
```

```python
# to fill all null values with median values of each category
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Category')['Review Rating'].transform('median'))
```

```python
df.isnull().sum()
```

```python
# changing all column names to lower case with underscore (snake casing)
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
age_group = ['Kids', 'Adults', 'Middle_aged', 'Senior']
df['age_groups'] = pd.qcut(df['age'], q = 4, labels = age_group)
```

```python
df[['age','age_groups']].head(10)
```

```python
# create a new column purchase_frequency_days
df[['frequency_of_purchases']]
```

```python
frequency_mapped = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Quartely' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Monthly' : 30,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapped)

```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']]
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
# if promocode used , discount always applied, so we dont need promocode column
df.drop('promo_code_used', axis = 1, inplace = True)
```
